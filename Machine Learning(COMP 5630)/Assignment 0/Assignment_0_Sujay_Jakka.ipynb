```python
#Problem 1

import numpy as np
import matplotlib.pyplot as plt

test_file = open("test.txt", "r")
file_content = test_file.read()

histogram_dict = {}
normalized_histogram_dict = {}

for c in file_content:
  if c.isalpha():
    c = c.lower()
    if c in histogram_dict:
      histogram_dict[c] += 1
    else:
      histogram_dict[c] = 1

histogram_dict = dict(sorted(histogram_dict.items()))
alphabet = np.array(list(histogram_dict.keys()))
histogram_np_array = np.array(list(histogram_dict.values()))

for c in histogram_dict:
  normalized_histogram_dict[c] = histogram_dict[c] / np.sum(histogram_np_array)

normalized_histogram_dict = dict(sorted(normalized_histogram_dict.items()))

print("This is the numpy array version of my histogram.")
print(histogram_np_array)

plt.title("Normalized Frequency V Alphabet")
plt.xlabel("Alphabet")
plt.ylabel("Normalized Frequency")
plt.bar(alphabet, histogram_np_array / np.sum(histogram_np_array))
plt.show()
print(normalized_histogram_dict)

plt.title("Frequency V Alphabet")
plt.xlabel("Alphabet")
plt.ylabel("Frequency")
plt.bar(alphabet, histogram_np_array)
plt.show()
print(histogram_dict)





```

```python
#Problem 2

import cv2

img = cv2.imread("test.png")
x, y, z = np.shape(img)

output_1 = np.zeros((x, y, z))
output_2 = img.copy()

for i in range(x):
  for j in range(y):
    vector = img[i][j]
    dist_1 = np.linalg.norm(vector - [255, 0, 0])
    dist_2 = np.linalg.norm(vector - [0, 255, 0])
    dist_3 = np.linalg.norm(vector - [0, 0, 255])

    min_dist = min(dist_1, dist_2, dist_3)

    if min_dist == dist_1:
      output_1[i][j] = [255, 0, 0]

    elif min_dist == dist_2:
      output_1[i][j] = [0, 255, 0]

    else:
      output_1[i][j] = [0, 0, 255]

cv2.imwrite("output1.png", output_1)

square_size = 50

x_start, y_start = (x - 50) // 2, (y - 50) // 2

output_2[x_start:x_start + 50, y_start:y_start + 50] = [0, 0, 0]

cv2.imwrite("output2.png", output_2)



```
